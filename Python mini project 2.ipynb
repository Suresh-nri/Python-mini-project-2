```python

def descending_order(arr):
    n = len(arr)
    i = 0
    while i < n:
        j = i + 1
        while j < n:
            if arr[i] < arr[j]:  # If the current element is smaller, rearrange
                temp = arr[i]
                arr[i] = arr[j]
                arr[j] = temp
            j += 1
        i += 1
    return arr

# Given list
numbers = [8, 10, 5, 4, 7, 23]

# Calling function
sorted_numbers = descending_order(numbers)

# Printing output
print(sorted_numbers)
```
